### 1. Define a function that returns a list of numbers that sum up to 1 

* Generate a list of random numbers
* Sum up all the numbers in the list
* Take each number in the initial list and divide it by the sum of all numbers
* Print the new list
* Sum up all the values in the new list. It should sum up to 1

```python
import random
random_numbers = [random.random() for i in range(20)]
total_sum = sum(random_numbers)
new_list = [i/total_sum for i in random_numbers]
print(new_list)
print(sum(new_list))
```

### My implementation
* Declare var max for the max sum of the array
* Declare var ini that will be the current sum of the array, initialized in zero because array is empty
* initialize array arr
* loops while the total sum of the array is less than the max sum
* inside the loop, generates a uniform random number between 0 and 0.1 exclusive
* appends the random number to the array
* increment the sum in r
* when ini>max,  breaks the loop and apend 1-sum(arr) verifying that the total sum will be 1.0
* prints the arr
* prints the sum

```python
import random
def genRandom():
    max = 1
    ini = 0
    arr = []
    while ini<max:
        r = random.uniform(0, 0.001)
        arr.append(r)
        ini+=r
    #arr[len(arr)-1]=arr[len(arr)-1]+(1-sum(arr))
    arr.append(1-sum(arr))
    print(arr)
    print(sum(arr))
    
    
for i in range(1):
    genRandom()
    
```
